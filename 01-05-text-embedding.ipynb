
text embeddingsのためにembedding-001モデルを使う

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

gemini_embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001") # type: ignore
     
```

```python
text = "Some normal text to send to OpenAI to be embedded into a N dimensional vector"
```

```python
embedded_text = gemini_embeddings.embed_query(text)
```

```python
print(type(embedded_text))
print(len(embedded_text))
print(embedded_text[0])
print(len(embedded_text))
```

```python
from langchain.document_loaders import CSVLoader
```

```python
path = '../langchain-course-main/June 2024 Update/01-Data-Connections/'
loader = CSVLoader(path+'some_data/penguins.csv')
```

```python
data = loader.load()
```
